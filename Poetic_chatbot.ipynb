**STEP 2: Import required Libraries**

```python
import time
from google import genai      #Gemini API
from google.colab import userdata     #For securely retrieving API key in Colab
```



---



**STEP 3: Get Gemini API Key Securely from Colab**

```python
key = userdata.get('Gemini_api')
```

**STEP 4: Function to Simulate Typing Animation**

This adds immersion by typing word-by-word with delay

```python
def stream(text):
  print('Bot:', end=' ')
  words = text.split()
  for word in words:
    print(word, end=' ', flush=True)
    time.sleep(0.1)
  print()
```

Testing the typing effect with a simple line.

```python
text = 'Good Morning!!'
stream(text)
```



---


**STEP 5: Initialize Gemini Client and Create Chat Instance**

```python
try:
  client = genai.Client(api_key=key)
  chat = client.chats.create(model="gemini-2.0-flash")
except KeyError as e:
  print(f'Error: Missing Gemini API key.\n{e}')
  raise SystemExit("Exiting due to missing API key.")
except Exception as e:
  print(f'Unexpected error while initializing Gemini client:\n{e}')
  raise SystemExit("Exiting due to client initialization failure.")
```



---



**STEP 6: Function to ask Gemini with Poetic Prompting**

```python
def ask_gemini(user_input):
  try:
    # Prompt Engineering: Instructing Gemini to behave like a poetic chatbot.
    prompt = f"You are a poetic chatbot. Respond to every question of user input with a poem that is related to the user input, even if it's a serious question. Keep it short and sweet! Here's the question {user_input}"
    response = chat.send_message(prompt)
    return response.text
  except Exception as e:
    return f"Oops, something went wrong: {e}"
```



---



**STEP 7: Chatbot Interaction Loop**

```python
print("Welcome to Chatbot!! Type 'quit' to exit")

while True:
  user_input = input("You: ")
  if user_input.lower() == 'quit':
    print("Bot: Bye for now!")
    break
  response = ask_gemini(user_input)
  stream(response)
```
